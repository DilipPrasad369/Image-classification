```python
from transformers import ViTImageProcessor, ViTForImageClassification
from PIL import Image
import requests
```

```python
import torch
```

```python
import os
```

```python
import pandas as pd
import numpy as np
```

```python
def classify(path):
  image = Image.open(path)

  # Convert GIF to JPEG
  if image.format == "GIF":
    image = image.convert("RGB")

  processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
  model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')

  # Convert PIL image to tensor
  inputs = processor(images=image, return_tensors="pt", format="torch")
  image_tensor = inputs.pixel_values

  inputs = processor(images=image, return_tensors="pt")

  outputs = model(**inputs)

  logits = outputs.logits

  predicted_class_idx = logits.argmax(-1).item()

  return model.config.id2label[predicted_class_idx]

```

```python
birds = os.listdir('/content/train')
```

```python
df = pd.read_csv('birds.csv')
```

```python
paths=[]

for bird in birds:
    bird_df = df[df['labels'] == bird]
    if not bird_df.empty:
      paths.append(bird_df['filepaths'].iloc[0])
    else:
        print(f"No data found for bird: {bird}")
```

```python
birds.remove('.ipynb_checkpoints')
```

```python
for i in range(len(birds)):
  y_pred = classify(path=paths[i])
  print('Passed->{} & Predicted->{}'.format(birds[i],y_pred))
```
